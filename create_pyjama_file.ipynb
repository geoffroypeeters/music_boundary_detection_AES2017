```python
import sys
sys.path.append('../')
import constants as const

import os
from pyjama import pyjama
import glob
import pandas as pd
import pprint as pp
```

# Remove previously existing file

```python
if os.path.exists('./rwc-pop.pyjama'):
    os.remove('./rwc-pop.pyjama')
```

# Create new pyjama file

```python
myPyjama = pyjama.C_pyjama(notValidAction='addToDictionary')
myPyjama.M_addDefinition(descriptionName='filepath', 
                            typeConstraint='free')
myPyjama.M_addDefinition(descriptionName='structtype',
                             typeExtent='segment',
                             typeConstraint='free',
                             generator={'name': '', 'version': '', 'date': ''})
```

```python
audio_file_l = glob.glob(f'{const.root_path}/_references/rwc-pop/audio/*.wav')
audio_file_l = sorted(audio_file_l)
for idx, audio_file in enumerate(audio_file_l):
    print(audio_file)
    root = audio_file.split('/')[-1].split('.')[0]
    annot_file = f'{const.root_path}/_references/rwc-pop/rwc-chorus/AIST.RWC-MDB-P-2001.CHORUS/{root}.CHORUS.TXT'
    data_pd = pd.read_csv(annot_file, sep='\t', header=None)
    data_np = data_pd.values
    
    myPyjama.M_addEntry()
    myPyjama.M_updateEntry(descriptionName='filepath', value_l=audio_file)
    nb_seg = data_np.shape[0]
    for num_seg in range(nb_seg):
        myPyjama.M_updateEntry(descriptionName='structtype', value_l=[data_np[num_seg,2]], time_l=data_np[num_seg,0]/100, duration_l=data_np[num_seg,1]/100-data_np[num_seg,0]/100)

myPyjama.M_save('./rwc-pop.pyjama')
```

```python
print(len(myPyjama.data['collection']['entry']))
pp.pprint(myPyjama.data['collection']['entry'][0])
```
